```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path
file_path = 'data/ScadaFlow_17-06-2024_To_18-06-2024-clean.xlsx'

# Define the specific columns for each sheet
columns_volume = ['ReportDate', 'Volumetric Flow G Previous Rate (SCFM) * 24', 'Volumetric Flow L Previous Rate (bbl/hr) *24', 'Volumetric Flow G App Previous Rate (SCFM) * 24']
columns_line = ['ReportDate', 'Line Pressure (psi)', 'Line Temperature (F)']
columns_h20 = ['ReportDate', 'D Pppl (InH2O)', 'D Pr (InH2O)', 'D Pt (InH2O)']

# Load the specific sheets with the defined columns
volume_df = pd.read_excel(file_path, sheet_name='Volume', usecols=columns_volume)
line_df = pd.read_excel(file_path, sheet_name='Line', usecols=columns_line)
h20_df = pd.read_excel(file_path, sheet_name='H20', usecols=columns_h20)

# Convert ReportDate to datetime
volume_df['ReportDate'] = pd.to_datetime(volume_df['ReportDate'])
line_df['ReportDate'] = pd.to_datetime(line_df['ReportDate'])
h20_df['ReportDate'] = pd.to_datetime(h20_df['ReportDate'])

# Plot comparison charts
plt.figure(figsize=(15, 10))

# Volume chart
plt.subplot(3, 1, 1)
plt.plot(volume_df['ReportDate'], volume_df['Volumetric Flow G Previous Rate (SCFM) * 24'], label='Volumetric Flow G (SCFM) * 24')
plt.plot(volume_df['ReportDate'], volume_df['Volumetric Flow L Previous Rate (bbl/hr) *24'], label='Volumetric Flow L (bbl/hr) *24')
plt.plot(volume_df['ReportDate'], volume_df['Volumetric Flow G App Previous Rate (SCFM) * 24'], label='Volumetric Flow G App (SCFM) * 24')
plt.title('Volume Comparison')
plt.xlabel('Report Date')
plt.ylabel('Flow Rate')
plt.legend()

# Line chart
plt.subplot(3, 1, 2)
plt.plot(line_df['ReportDate'], line_df['Line Pressure (psi)'], label='Line Pressure (psi)')
plt.plot(line_df['ReportDate'], line_df['Line Temperature (F)'], label='Line Temperature (F)')
plt.title('Line Comparison')
plt.xlabel('Report Date')
plt.ylabel('Value')
plt.legend()

# H20 chart
plt.subplot(3, 1, 3)
plt.plot(h20_df['ReportDate'], h20_df['D Pppl (InH2O)'], label='D Pppl (InH2O)')
plt.plot(h20_df['ReportDate'], h20_df['D Pr (InH2O)'], label='D Pr (InH2O)')
plt.plot(h20_df['ReportDate'], h20_df['D Pt (InH2O)'], label='D Pt (InH2O)')
plt.title('H20 Comparison')
plt.xlabel('Report Date')
plt.ylabel('InH2O')
plt.legend()

plt.tight_layout()
plt.show()
```
